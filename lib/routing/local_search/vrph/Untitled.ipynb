```python

import numpy as np
import importlib

import VRPH
importlib.reload(VRPH)
from VRPH import VRP
from utils.definitions import *

import tsplib95
from utils.plot_utils import plot_routes, plot_nodes
import matplotlib.pyplot as plt

from utils.data_utils import generate_tsp
from utils.VRPH_utils import load_tsp

n = 48

data = generate_tsp(n,1,seed = 1245)

#coords = np.insert(data[0]["coordinates"],0,[0,0]).reshape(-1,2)
coords = data[0]["coordinates"]
V = load_tsp(coords)




rnd = np.random.default_rng(123)
tour = rnd.permutation(n-1)+1
tour = np.append(tour, 0)

#C++ initial
#tour = np.array([21,15,40,33,13,12,24,47,4,28,1,41,25,3,34,44,9,23,31,38,20,46,19,32,35,29,42,16,26,18,36,5,27,6,17,43,30,45,14,11,10,22,2,39,37,8,7,0 ])

#V.call_clarke(1.2,False)
V.use_initial_solution([(tour).tolist()])
V.show_routes()

heuristics = TWO_OPT + THREE_OPT
rules = VRPH_BEST_ACCEPT + VRPH_FREE
m = [V.get_routes()[0]]
cost = [V.get_total_route_length()]
iters = 50
for i in range (iters):
    new_cost = V.detailed_solve(
        heuristics,  # local_operators
        rules,  # rule
        [-1],       # node (-1 for consider all nodes)
        0,       # steps
        1,  # iters
        1e-5,  # err_max
        False  # converge
    )
    #new_cost = model.Single_solve(heuristics, rules,  0.001,  1, False)
    m .append(V.get_routes()[0])
    #cost.append(V.get_total_route_length())
    cost.append(new_cost)


from ipywidgets import interact

def p(sol):
    fig, ax = plt.subplots(1,1,figsize=(10, 10))

    plot_nodes(ax, coords, data[0]['depot'])
    plot_routes(ax, m[sol], coords)
    plt.title(str(cost[sol]))
    plt.show()
    
interact(p, sol =(0, iters-1, 1))
```
